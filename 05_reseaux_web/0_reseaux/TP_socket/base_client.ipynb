**Consignes**: activité à faire en binôme: l'un avec le notebook «serveur» l'autre avec le notebook «client» (celui-ci)

```python
from socket import socket
```

    ---> attendre que votre binome soit en attente de connexion (étape1)

## étape 1 - se connecteur au serveur

Le serveur est en attente de connexion...

On crée un canal entre ce client et le serveur.

```python
s = socket()
s.connect(('127.0.0.1', 11111))
```

    ---> aller à l'étape 2 du programme serveur

## étape 2 - envoyer des données au serveur

La connexion est établie, il temps d'envoyer un message au serveur et de se mettre en attente d'une réponse éventuelle:

```python
# remplacer ___ par votre message (non vide!)
message = b'mets y du tiens un peu' 
s.send(message)
# en attente
recu = s.recv(1024)
print(recu)
```

    ---> attendre que votre binôme vous envoie des données (étape3)
    
    ---> dès que la cellule se débloque, modifier le message et exécuter à nouveau (autant de fois que voulu puis passer à la dernière étape)

## étape 3 - fermer la connexion

le client veut rompre la connexion, il doit prévenir le serveur avec un `.close()`.

Si le serveur est en attente de données, il reçoit un message vide qui lui permet d'agir en conséquence:

généralement, il va fermer son canal de communication vers le client et:
- soit se remettre en attente d'une nouvelle connexion entrante (en général),
- soit fermer son canal de communication avec l'interface réseau. 

```python
s.close()
```
