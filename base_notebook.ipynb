# Gestion des cellules d'un notebook

Voir cette [vidéo](https://vimeo.com/455065259)

- **insérer** une cellule avant `A` (Above), après `B` (Below)

- **supprimer** une cell.: `D D`

- **exécuter** une cell.: `CTRL+ENTER` ou `MAJ+ENTER`.

- **changer le mode** d'une cellule: `M` pour Markdown, `Y` pour code, `R` pour raw  (brut) - moins utile

- **sélection multiple** avec `MAJ` et cliquer.

- **scinder** une cellule en deux autres: `CTRL+MAJ+D`

- **fusionner** deux cellules: `MAJ+M`

- **copier** `C`, **couper** `X`, **coller** `V`.

- **annuler** une commande: `Z`, **annuler l'annulation précédente** `MAJ Z` 

- **supprimer la sortie** d'une cellule de code: `D O`

- **interrompre** une cellule qui n'en finit pas: `I I`.
